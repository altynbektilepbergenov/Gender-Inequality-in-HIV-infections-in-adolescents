**Recreating a graph about estimated number of people living with HIV and how it changed over the some period of time(years) **

```python
df['Year'].value_counts()
```

```python

df.groupby('Year')['Estimated number of people living with HIV'].nunique().plot(kind='line',c='r',marker=10);
plt.title('Change in estimated number of people with HIV',fontsize = 20,c='blue')
plt.xlabel('Year',fontsize = 16)
plt.ylabel('Estimated number of people living with HIV',fontsize = 16)
plt.style.use('ggplot')
plt.grid(c = 'g', alpha = .9,linestyle = '-')
plt.show()


```

As we can see that the overall movement of change in estimated number of people were increased from 1990 to 2020, and in detail it was fluctuated over years. 

Interestingly,number of people with HIV in 2015 and 2020 are almost the same. 

Between 2010 and 2015, the rate of number of people were consistent(no change, in plateau position). 

Overall, 1990 year has the least quantity of infected ones, 2020 year has the most quantity of infected ones

Probably, in my opinion, the change in number of infected people are progressive and will be increased in upcoming years

**2.Plotting a graph on estimated number of annual new infections and how it moderated over years**

```python
df['Year'].nunique()
```

```python
df['Year'].count()
```

```python
df['Year'].size
```

```python
df.loc[df['Estimated number of annual new HIV infections']=='100','Year'].agg(['nunique','count','size'])
```

```python

df.groupby('Year')['Estimated number of annual new HIV infections'].nunique().plot(kind='line',c='orange',marker=10)
plt.title('Change in estimated number of annual new HIV infections',fontsize = 18,c='black')
plt.xlabel('Year',fontsize = 16)
plt.ylabel('Estimated number of annual new HIV infections',fontsize = 14)
plt.style.use('dark_background')
plt.grid(c = 'white', alpha = .9,linestyle = '-')
plt.show()
```

**Conclusions:**

The overall trend in the number of new HIV infections is dinamic

In 1990, the number of new infections was quite low, then it increased significantly by 1993, after that it dropped slightly by small amount in 1995

In 2017, overall change started to fall down gradually, and it reached its lowest point in between 2017 and 2020

In my opinion, if there are any new HIV infections in following years in future, then the number of new infected will be lowered and they can be infected only by other factors




**3.Change in between death rate and Sex of population over years **



```python
import seaborn as sns
import matplotlib.patches as mpatches
```

```python
sns.lineplot(x='Year', y='Estimated number of annual AIDS related deaths',data=df, hue='Sex', palette=['red','green']);
plt.title('Death rate change in years by sex',c='black',fontsize=18)
plt.xlabel('Year',fontsize = 16,c='red')
plt.ylabel('Estimated number of annual AIDS related deaths',fontsize = 16, c='blue')
plt.style.use('classic')
plt.grid(c = 'black', alpha = .9,linestyle = '-')
red_patch = mpatches.Patch(color='red',label='Female')
green_patch=mpatches.Patch(color='green',label='Male')
plt.legend(loc=(1.02,0),handles=[red_patch,green_patch])


plt.show()
```

Conclusion:
Death among females are higher than among the males in 2010

In 1990, the female death rate was about 150 people , the male death rate at aprroximate below 100 people

The highest death rate of females was above 600 people per year in 2010, the highest death rate of males was over 550 people in 2016,2017

Overall, the trend in change of death rate was ascending in first 20 years, then it started to descend gradually until 2020, and I can even predict that this trend will remain low
